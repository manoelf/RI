```python
INVERTED_INDEX = dict()
```

```python
#seach term and calculate tf (term-frequency)

def search_term(term, data):
    result = []
    cont = 0
    
    if (term in INVERTED_INDEX):
        result = INVERTED_INDEX[term]
    else:
        
        rows = data.shape[0]
                
        for doc in range(rows):
            title = (data.loc[doc, 'titulo']).lower()
            sub_title = (data.loc[doc, 'subTitulo']).lower()
            content = (data.loc[doc, 'conteudo']).lower()
                        
            tf = 0
            text = nltk.word_tokenize(title + ' ' + sub_title + ' ' + content)
            i = 0
            
            while (i < len(text)):
                if (text[i].lower() == term):
                    tf += 1
                    exist = True
                i += 1
            if (tf):
                id_notice = data.loc[doc, 'idNoticia']
                result.append((id_notice, tf))
                cont += 1
        INVERTED_INDEX[term] = (cont, result)
    
    return INVERTED_INDEX[term]
```

```python
def intersection(term1, term2): #Just the dosc that really matter, the ones that the term appen in both
    result = []
    

    i, j = 0, 0
    term1_length = term1[0]
    term2_length = term2[0]
    docs_1 = term1[1]
    docs_2 = term2[1]
    while (i < term1_length and j < term2_length):
        if (docs_1[i][0] == docs_2[j][0]):
            result.append(docs_1[i])
            
            i += 1
            j += 1
        elif (docs_1[i][0] < docs_2[j][0]):
            i += 1
        else:
            j += 1
    
    return result
    
    
```

```python
def vsm_binary(query, data):
    q = query.split()
    dic_docs = dict()
    
    for term in q:
        term_set = search_term(term, data)
        for doc in term_set[1]:
            if (doc[0] in dic_docs):
                dic_docs[doc[0]].append(doc[1])
            else:
                dic_docs[doc[0]] = [doc[1]]
    result = []
    
    for doc in dic_docs.keys():
        if (len(dic_docs[doc]) == len(q)):
            result.append((doc, len(q)))
            
    return result
        

    
    
```

```python
result = vsm_binary('compra de voto', df)
result.sort(key=operator.itemgetter(1), reverse=True)
print(result[:5])
```

```python
def vsm_tf(query, data):
    q = query.split()
    dic_docs = dict()
    
    for term in q:
        term_set = search_term(term, data)
        for doc in term_set[1]:
            if (doc[0] in dic_docs):
                dic_docs[doc[0]].append(doc[1])
            else:
                dic_docs[doc[0]] = [doc[1]]
    result = []
    
    for doc in dic_docs.keys():
        if (len(dic_docs[doc]) == len(q)):
            result.append((doc, sum(dic_docs[doc])))
            
    return result
```

```python
result = vsm_tf('compra de voto', df)
result.sort(key=operator.itemgetter(1), reverse=True)
print(result[:5])
```

```python
def vsm_tf_idf(query, data):
    q = query.split()
    dic_docs = dict()
    
    for term in q:
        term_set = search_term(term, data)
        for doc in term_set[1]:
            if (doc[0] in dic_docs):
                dic_docs[doc[0]].append(doc[1])
            else:
                dic_docs[doc[0]] = [doc[1]]
    result = []
    
    for doc in dic_docs.keys():
        tf_idf_sum = 0
        if (len(dic_docs[doc]) == len(q)):
            for i in range(len(q)):
                tf_idf_sum += dic_docs[doc][i] * idf(q[i], data)
            result.append((doc, tf_idf_sum))
            
    return result





def idf(term, data):
    size = search_term(term, data)[0]
    total_docs = data.shape[0]
    calc = math.log((total_docs - 1)/ size)
    
    return calc
```

```python
result = vsm_tf_idf('compra de voto', df)
result.sort(key=operator.itemgetter(1), reverse=True)
print(result[:5])
```

```python
def vsm_bm25(query, data):
    q = query.split()
    dic_docs = dict()
    
    for term in q:
        term_set = search_term(term, data)
        for doc in term_set[1]:
            if (doc[0] in dic_docs):
                dic_docs[doc[0]].append(doc[1])
            else:
                dic_docs[doc[0]] = [doc[1]]
    result = []
    
    for doc in dic_docs.keys():
        bm25_sum = 0
        if (len(dic_docs[doc]) == len(q)):
            for i in range(len(q)):
                bm25_sum += bm25(dic_docs[doc][i])
            result.append((doc, bm25_sum))
            
    return result
    
    
def bm25(tf):
    k = 5
    calc = ((k + 1) * tf)/(tf + k)
    return calc
```

```python
result = vsm_bm25('compra de voto', df)
result.sort(key=operator.itemgetter(1), reverse=True)
print(result[:5])
```
